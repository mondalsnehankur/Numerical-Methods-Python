# ***Monte Carlo Integration.***

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy.integrate import simps
```

```python
def f(x):
  return 0.5*np.exp(-x)+np.sin(2*x)
```

```python
a,b,n, = 0,1,100
x = np.linspace(a,b,n)
y = [f(i) for i in x]

xmin = x[0]
xmax = x[n-1]

ymin = 0
ymax = 1.02 * np.max(y)

sidex = xmax - xmin
sidey = ymax - ymin

rect_area = sidex*sidey
```

```python
yup = []
ydown = []
xup = []
xdown = []

trial = 10000
count = 0
```

```python
for i in range(trial):
  xr = xmin + sidex*random.random()
  yr = ymin + sidey*random.random()

  if yr<=f(xr):
    count += 1
    ydown.append(yr)
    xdown.append(xr)
  else:
    yup.append(yr)
    xup.append(xr)

frac_area = rect_area*float(count)/trial
print('Integration :',frac_area)
```

```python
# PLOTTING

plt.plot(x,y,'-',color='k',lw = 2)
plt.plot(xup,yup,'.',color='k')
plt.plot(xdown,ydown,'+',color='grey')

plt.axhline(y=ymax,linestyle='--',color='k')
plt.axhline(y=ymin,linestyle='--',color='k')
plt.axvline(x=xmax,linestyle='--',color='k')
plt.axvline(x=xmin,linestyle='--',color='k')

plt.title('$Monte$ $Carlo$ $Integraction$.')
plt.grid()
plt.show()
```
