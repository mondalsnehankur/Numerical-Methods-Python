```python
import numpy as np
from scipy.integrate import quad
```

```python
# Define the wave function for the ground state
def psi(x):
    L = 2
    return np.sqrt(1 / L) * np.sin(np.pi * x / L)

# Define the integrands for the expectation values
def integrand_x(x):
    return x * (psi(x) ** 2)

def integrand_2_minus_x(x):
    return (2 - x) * (psi(x) ** 2)

def integrand_x_times_2_minus_x(x):
    return x * (2 - x) * (psi(x) ** 2)
```

```python

# Calculate the expectation values
x_expectation, _ = quad(integrand_x, 0, 2)
expectation_2_minus_x, _ = quad(integrand_2_minus_x, 0, 2)
expectation_x_times_2_minus_x, _ = quad(integrand_x_times_2_minus_x, 0, 2)

# Print the results
print(f"Expectation value ⟨x⟩: {x_expectation:.5f}")
print(f"Expectation value ⟨2 - x⟩: {expectation_2_minus_x:.5f}")
print(f"Expectation value ⟨x(2 - x)⟩: {expectation_x_times_2_minus_x:.5f}")
```
