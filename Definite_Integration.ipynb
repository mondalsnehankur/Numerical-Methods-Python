***Method 1:*** Gauss Quadrature method

```python
import numpy as np
from scipy.integrate import quad
```

```python
# Define the function to integrate
def f(x):
    return 1/(x**2 + 4)

# Perform definite integration from 0 to 1
result, error = quad(f, 0, 3)
print("Result:", f"{result:.3f}")
#print("Estimated error:", error)
```

***Method 2:*** Using Trapezoidal method

```python
#!pip install --upgrade scipy

import numpy as np
#from scipy.integrate import trapz
```

```python
x = np.linspace(0, 3, 100)
y = 1/(x**2 + 4)  # Corresponding function values

# Perform definite integration using trapezoidal rule
result = np.trapz(y, x)
print("Result:", f"{result:.3f}")
```

***Method 3:*** Using Simpsons method

```python
#!pip install --upgrade scipy

import numpy as np
from scipy import integrate
```

```python
# Sample data points
x = np.linspace(0, 3, 100)
y = 1/(x**2 + 4) # Corresponding function values

# Perform definite integration using Simpson's rule
result = integrate.simps(y, x)
print("Result:", result)
```

***Method 4:*** Symbolic integration.

```python
import sympy as sp

# Define the variable and function
x = sp.symbols('x')
f = 1/( x**2 + 4)

# Perform definite integration from 0 to 1
result = sp.integrate(f, (x, 0, 3))
print("Result:", result)
print('Value of the Definite Integral :', f"{result.evalf():.4f}")
```
